```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

import warnings
warnings.simplefilter('ignore')
```

```python
df = pd.read_csv('train.csv')
df
```

### Побудуємо діаграми розподілу для значень з поля 'Age' датасету train

```python
not_nan=df[df['Age'].notna()]
```

```python
sns.displot(not_nan, x='Age', kde=True)
plt.axvline(x=not_nan.Age.mean(), color='red', label='Mean')
plt.axvline(x=not_nan.Age.median(), color='blue', label='Median')
plt.legend(loc='upper right')
print(not_nan.Age.mean())
print(not_nan.Age.median())
```

### Заповнимо пропущені значення в датасеті

  #### Метод заповнення середнім значенням (по стовпчику)

```python
df['Age'].fillna(df['Age'].mean(), inplace = True)
```

 #### Метод заповнення значенням, яке  найчастіше зустрічається (в конкретному стовпчику)


```python
df['Embarked'].replace(np.nan, 'S', inplace = True)
```
